```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.pandas.set_option("display.max_columns", None)
```

```python
dataset = pd.read_csv("advance_house_price_prediction_train.csv")

dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset,dataset["SalePrice"],test_size = 0.1)
```

```python
nan_features = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes  == "O"]

for feature in nan_features:
    print(feature, "has", np.round(dataset[feature].isnull().mean()*100,4), "% missing values.")
```

```python
def replace_nan_values(dataframe, missing_features):
    dataframe[missing_features] = dataframe[missing_features].fillna("Missing")
    return dataframe

dataset = replace_nan_values(dataset, nan_features)

dataset[nan_features].isnull().sum()
```

```python
dataset.head()
```

```python
num_nan_features = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1 and dataset[feature].dtypes != "O"]
num_nan_features
```

```python
for feature in num_nan_features:
    print(f"{feature} has {np.round(dataset[feature].isnull().mean()*100,4)} % NaN values.")
```

```python
for feature in num_nan_features:
    median_value = dataset[feature].median()
    dataset[feature+"nan"] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value, inplace = True)
dataset[num_nan_features].isnull().sum()
```

```python
dataset.head()
```

```python
for feature in ["YearBuilt", "YearRemodAdd", "GarageYrBlt"]:
    dataset[feature] = dataset["YrSold"]-dataset[feature]
```

```python
dataset.head()
```

```python
num_features = ["LotFrontage", "LotArea", "1stFlrSF", "GrLivArea", "SalePrice"]

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == "O"]
categorical_features
```

```python
for feature in categorical_features:
    temp = dataset.groupby(feature)["SalePrice"].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature],"Rare_Var")
```

```python
dataset.head(100)
```

```python
encoded_features = pd.get_dummies(dataset[categorical_features])
```

```python
dataset.drop(categorical_features,inplace=True, axis=1)
```

```python
dataset = pd.concat([dataset, encoded_features], axis = 1)
```

```python
dataset.head()
```

```python
feature_scaled = [feature for feature in dataset.columns if feature not in ["Id", "SalePrice"]]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scaled])
```

```python
df = scaler.transform(dataset[feature_scaled])
```

```python
data = pd.concat([dataset[["Id", "SalePrice"]].reset_index(drop = True),
                     pd.DataFrame(df, columns = feature_scaled)], axis = 1)
```

```python
data.head()
```

```python
data.to_csv("remodelled data for advance house price prediction train.csv", index = False)
```
