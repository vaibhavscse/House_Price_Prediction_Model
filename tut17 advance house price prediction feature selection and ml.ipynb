```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option("display.max_columns", None)
```

```python
dataset = pd.read_csv("remodelled data for advance house price prediction train.csv")
```

```python
dataset.head()
```

```python
y_train = dataset[["SalePrice"]]
```

```python
x_train = dataset.drop(["Id","SalePrice"], axis = 1)
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state = 0))
# Note - The bigger the alpha value the less will the features be selected
feature_sel_model.fit(x_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_features = x_train.columns[feature_sel_model.get_support()]
print("No of selected features is",len(selected_features))
print("Total number of features is",len(x_train.columns))
```
