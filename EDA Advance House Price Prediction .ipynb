# Missing Values

```python
features_with_nan_values = [features for features in dataset.columns if dataset[features].isnull().sum()>=1]
```

```python
for feature in features_with_nan_values:
    print(feature,"-", np.round(dataset[feature].isnull().mean()*100, 4), "% missing values")
```

```python
for feature in features_with_nan_values:
    data = dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(), 1 , 0)
    
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != "O"]
# in the line aboive capital "O" represents object data type(i.e. non string data type)
print("The number of numerical features are", len(numerical_features))

dataset[numerical_features].head()
```

```python
year_feature = [feature for feature in dataset.columns if "Yr" in feature or "Year" in feature]

year_feature
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price Vs Year Sold")
```

```python
year_feature
```

```python
colors = ['blue', 'red', 'green']
for i, feature in enumerate(year_feature):
    if feature != "YrSold":
        data = dataset.copy()
        data[feature] = data["YrSold"]-data[feature]
        
        plt.scatter(data[feature], data['SalePrice'],c = colors[i%len(colors)])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+["Id"]]
discrete_features
```

```python
for feature in discrete_features:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.show()
```

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_features+year_feature+["Id"]]
continuous_features
```

```python
for feature in continuous_features:
    data = dataset.copy()
    data[feature].hist(bins = 30)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
data = dataset.copy()
for feature in continuous_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]= np.log(data[feature])
        data["SalePrice"] = np.log(data["SalePrice"])
        plt.scatter(data[feature], data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(feature+" vs SalePrice")
        plt.show()
```
